# GRAFIK PIE

```python
data_pie = df["Skill Moves"].value_counts()
data_pie.head()
```

```python
data_pie = df["Preferred Foot"].value_counts().rename_axis("Skill Moves")
data_pie.head()
```

```python
data_pie = df["Skill Moves"].value_counts().rename_axis("Skill Moves").reset_index(name="id_count")
data_pie.head()
```

```python
plt.title("Grafik Pie")
plt.pie(data_pie["id_count"],labels=data_pie["Skill Moves"],startangle=90,autopct="%.2f%%",shadow=True)
plt.show()
```
